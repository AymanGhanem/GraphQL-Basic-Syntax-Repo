```python
import pandas as pd
import random
random.seed(10)
rating_labels = ["ONE_STAR", "TWO_STARS", "THREE_STARS", "FOUR_STARS", "FIVE_STARS"]
```

```python
ids      = ['mov-' + str(i) for i in range(1,11)]
titles   = ["Rush Hour 1", "Rush Hour 2", "Rush Hour 3", "The Invisible Guest", "In Time", "Now You See ME", \
            "The Equalizer", "The Bone Collector", "The Lincolyn Lawyer", "Shutter Isaland"
           ]
actors   = ["Jackie Chan, Chris Tucker", "Jackie Chan, Chris Tucker", "Jackie Chan, Chris Tucker", \
            "Mario Casas, Ana Wagener", "Justin Timberlake, Olivia Wilde", "Jesse Eisenberg, Mark Ruffalo", \
            "Denzel Washington, Melissa Leo", "Denzel Washington, Angelina Jolie",\
            "Matthew McConaughey, Ryan Phillippe", "Leonardo DiCaprio, Mark Ruffalo"
           ]
availables = [random.choice([True, False]) for _ in range(10)]
ratings = [random.choice(rating_labels) for _ in range(10)]
movies_table = pd.DataFrame({"id": ids, "title": titles, "actors": actors, "available": availables, \
                             "rating": ratings
                            })
```
