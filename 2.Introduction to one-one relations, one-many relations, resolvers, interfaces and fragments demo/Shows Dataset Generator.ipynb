```python
import pandas as pd
```

```python
import datetime
import random
random.seed(30)
```

```python
dates = [datetime.datetime.today() + datetime.timedelta(days=random.randint(100,356)) for _ in range(20)]
dates = [str(date.day) + "-" + str(date.month) + "-" + str(date.year) for date in sorted(dates)]
names  = ["Rush Hour 1", "Rush Hour 2", "Rush Hour 3", "The Invisible Guest", "In Time", "Now You See ME", \
            "The Equalizer", "The Bone Collector", "The Lincolyn Lawyer", "Shutter Isaland"
        ]
names = [random.choice(names) for _ in range(20)]
tickets = [name + "*" +  str(random.choice(['VIP','Ordinary'])) + "*" + date + "-ticket" for name, date in zip(names, dates)]
```

```python
df = pd.DataFrame({'dates': dates, 'names': names})
shows_tickets = []
for row, col in df.iterrows():
    date = df.iloc[row][0]
    name = df.iloc[row][1]
    tickets = ','.join([name + "*" + random.choice(["vip", "ord"]) + "*" + date for _ in range(random.randint(1,100))])
    shows_tickets.append(tickets)
```

```python
df['tickets'] = shows_tickets
```
